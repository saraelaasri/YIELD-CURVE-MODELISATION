# spline cubique:
Etape1: Creation du jacobienne

```python
def jacobi(A, b, x0, tol, n_iterations=100):
   
    
    n = A.shape[0]
    x = x0.copy()
    x_prev = x0.copy()
    counter = 0
    x_diff = tol+1
    
    while (x_diff > tol) and (counter < n_iterations): #iteration level
        for i in range(0, n): #element wise level for x
            s = 0
            for j in range(0,n): #summation for i !=j
                if i != j:
                    s += A[i,j] * x_prev[j] 
            
            x[i] = (b[i] - s) / A[i,i]
        #update values
        counter += 1
        x_diff = (np.sum((x-x_prev)**2))**0.5 
        x_prev = x.copy() #use new x for next iteration
        
    return x
```

```python
def cubic_spline(x, y, tol = 1e-1000):
    """
    b, coefficient de x  degree 1
    c, coefficient de x  degree 2
    d, coefficient de x  degree 3
    """ 
    x = np.array(x)
    y = np.array(y)
   
    if np.any(np.diff(x) < 0):
        idx = np.argsort(x)
        x = x[idx]
        y = y[idx]
    size = len(x)
    delta_x = np.diff(x)
    delta_y = np.diff(y) 
    ### matrice A
    A = np.zeros(shape = (size,size))
    b = np.zeros(shape=(size,1))
    A[0,0] = 1
    A[-1,-1] = 1
    for i in range(1,size-1):
        A[i, i-1] = delta_x[i-1]
        A[i, i+1] = delta_x[i]
        A[i,i] = 2*(delta_x[i-1]+delta_x[i])
    ### Get matrix b
        b[i,0] = 3*(delta_y[i]/delta_x[i] - delta_y[i-1]/delta_x[i-1])    
    ### Solver c de l'equation Ac = b
    c = jacobi(A, b, np.zeros(len(A)), tol = tol, n_iterations=1000)
    ### trouver  d et b
    d = np.zeros(shape = (size-1,1))
    b = np.zeros(shape = (size-1,1))
    for i in range(0,len(d)):
        d[i] = (c[i+1] - c[i]) / (3*delta_x[i])
        b[i] = (delta_y[i]/delta_x[i]) - (delta_x[i]/3)*(2*c[i] + c[i+1])    
    
    return b.squeeze(), c.squeeze(), d.squeeze()
```

```python
def spline(a,m,l,maturity,f):
    r=[]
    for i in range(len(maturity)):
        if maturity[i]<=5:
            r.append(f[0]+a[i]*maturity[i]+m[i]*pow(maturity[i],2)+
                     l[i]*pow(maturity[i],3))
            
        elif( maturity[i]<=10) &(maturity[i]>5) :
            r.append(f[0]+a[i]*maturity[i]+m[i]*pow(maturity[i],2)+
                     l[i]*pow(maturity[i],3)+
                     (a[i]-a[i-1])*pow(maturity[i]-5,3))
        elif (maturity[i]<=20) & (maturity[i]>10):  
            r.append(f[0]+a[i]*maturity[i]+
                     m[i]*pow(maturity[i],2)+l[i]*pow(maturity[i],3)+
                     (a[i-1]-a[i-2])*pow(maturity[i]-5,3)+
                     (a[i]-a[i-1])*pow(maturity[i]-10,3))
            
    return r
```

```python
def NSS(m,vi):
    k=0
    k=vi[0]+vi[1]*((1-np.exp(-m/vi[4]))/(m/vi[4]))+vi[2]*((1-np.exp(-m/vi[4]))/(m/vi[4])-np.exp(-m/vi[4]))
    
    return k
```

```python
def NSSA(m,vi):
    k=0
    k=vi[0]+vi[1]*((1-np.exp(-m/vi[4]))/(m/vi[4]))
    + vi[2]*((1-np.exp(-m/vi[4]))/(m/vi[4])-np.exp(-m/vi[4]))
    + vi[3]*((1-np.exp(-m/vi[5]))/(m/vi[5])-np.exp(-m/vi[5]))
    return k
```

```python
def list_taux(m,vi):
    taux_=[]
    for i in m:
        t= NSSA(i,vi)
        taux_.append(t)
    return taux_
```

```python
#on va creer une liste des taux
def list_taux1(m,vi):
    taux_=[]
    for i in m:
        t= NSS(i,vi)
        taux_.append(t)
    return taux_
#on trouve l'erreur
def obj1(facteur):
    e=(list_taux1(data['period'],facteur)-taux)**2
    scr= np.sum(e)
    return scr
#on cherche les coefficiants optimiser
resultat= minimize(obj1,vi)
obj_fct= resultat.fun
vo=resultat.x
print('objectif :',obj_fct , "\n"
     'valeurs opti :',vo)
```

```python
def cir(r0, NO,N, seed=777):
    
    np.random.seed(seed)
    dt = NO[2]/float(N) 
    rates=[r0]
    for i in range(N-1):
        dr = NO[0]*(NO[1]-rates[-1])*dt +NO[3]*math.sqrt(abs(rates[-1]))*np.random.normal()
        rates.append(rates[-1] + dr)
    return rates
```
